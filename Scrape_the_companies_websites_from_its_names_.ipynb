```python
import requests
from bs4 import BeautifulSoup
import csv

# Set the URL of a search engine
url = "https://www.google.de/search"

# Set the user agent to mimic a web browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.66 Safari/537.36"
}

# Open the CSV file containing the list of companies
with open("companies_liz.csv", "r") as file:
    # Read the CSV file using the csv module
    reader = csv.reader(file)

    # Open a new CSV file to write the results to
    with open("results.csv", "w", newline="") as result_file:
        # Create a CSV writer
        writer = csv.writer(result_file)

        # Loop through the rows in the CSV file
        for row in reader:
            # Get the name of the company
            company = row[0]

            # Set the search parameters
            params = {
                "q": company,
            }

            # Send the request to the search engine
            response = requests.get(url, params=params, headers=headers)

            # Check the status code to see if the request was successful
            if response.status_code == 200:
                # If the request was successful, parse the HTML of the search results page
                soup = BeautifulSoup(response.text, "html.parser")

            # Find the first result on the page
            result = soup.find("div", class_="g")

            # If a result was found, extract the website URL
            if result:
                website_url = result.find("a").get("href")
                print(f"{company}: {website_url}")
                # Write the company name and website URL to the CSV file
                writer.writerow([company, website_url])
            else:
                # If no result was found, write a message to the CSV file
                writer.writerow([company, "No website found"])

```
